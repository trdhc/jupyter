```python
import datetime as dt
import pandas as pd
from random import random, randint
```

```python
def generate_table(length=1000):
    
    Years = [randint(2013, 2018) for i in range(length)]
    Months = [randint(1, 12) for i in range(length)]
    Pairs = zip(Years, Months)
    
    Dates = [dt.date(pair[0], pair[1], 1) for pair in Pairs]
    
    Date_Stamps = [dt.datetime.strftime(date, '%Y-%m') for date in Dates]
    
    aF = pd.DataFrame([{'a':random(), 'b':random(), 'c':random()} for i in range(length)])
    
    aF = aF.assign(date_stamp=Date_Stamps)
    
    aG = aF.groupby('date_stamp')
    return(aG.describe())
```

```python
generate_table(1000000)
```
